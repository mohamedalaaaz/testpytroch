```python
import json
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments
from datasets import load_dataset, Dataset
from trl import SFTTrainer
import os

def parse_feedback(log_file):
    dataset = []
    with open(log_file, "r", encoding="utf-8") as f:
        entries = f.read().strip().split("\n\n")
        for entry in entries:
            lines = entry.strip().split("\n")
            if len(lines) < 3:
                continue
            q = lines[0].replace("Q: ", "").strip()
            a = lines[1].replace("A: ", "").strip()
            fb = lines[2].replace("Feedback: ", "").strip()

            # Skip empty or bad feedback
            if fb.lower() in ["", "pending", "n", "no", "bad", "wrong"]:
                continue

            dataset.append({"instruction": q, "input": "", "output": a})

    return dataset

def save_jsonl(dataset, output_file):
    with open(output_file, "w", encoding="utf-8") as f:
        for item in dataset:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")

def fine_tune(dataset_path, model_name="mistralai/Mistral-7B-Instruct"):
    tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
    model = AutoModelForCausalLM.from_pretrained(model_name, trust_remote_code=True)

    tokenizer.pad_token = tokenizer.eos_token

    dataset = load_dataset("json", data_files=dataset_path, split="train")

    training_args = TrainingArguments(
        output_dir="./mistral-finetuned-feedback",
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        num_train_epochs=2,
        logging_steps=10,
        save_strategy="epoch",
        learning_rate=2e-5,
        fp16=True,
        save_total_limit=2,
    )

    trainer = SFTTrainer(
        model=model,
        train_dataset=dataset,
        tokenizer=tokenizer,
        args=training_args,
        formatting_func=lambda x: {
            "text": f"<|user|>: {x['instruction']}\n<|assistant|>: {x['output']}"
        }
    )

    trainer.train()

if __name__ == "__main__":
    LOG_FILE = "feedback_log.txt"
    JSONL_FILE = "fine_tune_dataset.jsonl"

    if os.path.exists(LOG_FILE):
        data = parse_feedback(LOG_FILE)
        if data:
            save_jsonl(data, JSONL_FILE)
            print(f"Saved {len(data)} entries to {JSONL_FILE}")
            fine_tune(JSONL_FILE)
        else:
            print("No valid feedback found.")
    else:
        print(f"{LOG_FILE} not found.")
```

```python
from huggingface_hub import HfApi, Repository

def push_to_hub(output_dir="./mistral-finetuned-feedback", repo_id="YOUR_USERNAME/spacelion-model", token=os.getenv("HF_TOKEN")):
    api = HfApi()
    repo_url = api.create_repo(repo_id, exist_ok=True, token=token)
    repo = Repository(local_dir=output_dir, clone_from=repo_url, use_auth_token=token)
    repo.push_to_hub(commit_message="Update fine-tuned Spacelion model")
    print(f"‚úÖ Model pushed to Hugging Face Hub: {repo_id}")

# After trainer.train()
trainer.train()
push_to_hub()
```

```python
import streamlit as st
import os

FEEDBACK_FILE = "feedback_log.txt"
TRAIN_LOG = "training.log"

st.title("ü¶Å Spacelion Feedback & Training Dashboard")

if os.path.exists(FEEDBACK_FILE):
    with open(FEEDBACK_FILE, "r", encoding="utf-8") as f:
        feedback = f.read().strip().split("\n\n")
    st.write(f"### Total feedback entries: {len(feedback)}")
    st.write("### Recent feedback:")
    for entry in feedback[-5:]:
        st.write(entry)
else:
    st.write("No feedback logs found.")

if os.path.exists(TRAIN_LOG):
    st.write("### Training log (last 20 lines):")
    with open(TRAIN_LOG) as f:
        lines = f.readlines()[-20:]
    st.text("".join(lines))
else:
    st.write("No training logs found.")
```
